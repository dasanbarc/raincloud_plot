```python
# HIDE CODE

import ptitprince as pt
import matplotlib.pyplot as plt
import pandas as pd

# Creo dataframe:
archivo_csv = r"..\datasets\medical_dataset_RAIN_CLOUD_PLOT\insurance.csv"
archivo_csv = archivo_csv.replace("\\", "/")
df = pd.read_csv(archivo_csv, delimiter=",")
df = df.rename(columns={
    "bmi": "BMI",
    "region": "Region"
})

# Raincloud plot:
plt.figure(figsize=(8, 6))
ax = pt.RainCloud(
    x="Region",
    y="BMI",
    data=df,
    palette="Set1",
    bw=.2,
    width_viol=.9,
    width_box=.3,
    orient="h",
    jitter=1
)
plt.title("Raincloud Plot: Body Mass Index (BMI) by Region")
plt.show()
```
